```python
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.collections import PatchCollection as plt_circles

plt.rcParams['figure.dpi'] = 200
plt.rcParams['savefig.dpi'] = 200
```

```python
class Particles():

    particle_list = list()
    
    L = 1.0
    V0 = 1.0
    dt = 0.005

    def __init__(self,r):
        Particles.particle_list.append(self)
        self.r = r
        self.A = np.pi*self.r**2
        self.s = self.A*4/np.pi
        
        self.x = np.random.uniform(0,1)*self.L
        self.y = np.random.uniform(0,1)*self.L

        theta = np.random.uniform(0,2*np.pi)
        self.Vx = Particles.V0*np.cos(theta)
        self.Vy = Particles.V0*np.sin(theta)
        

    def step(self):
        self.x = self.x + self.Vx*Particles.dt
        self.y = self.y + self.Vy*Particles.dt

    def check_colision(self):
        pass

    def check_wall(self):
        pass


    @classmethod
    def plot_particles(cls):
        fig, ax = plt.subplots()
        ax.set_box_aspect(1)
        ax.set_xlim([0,cls.L])
        ax.set_ylim([0,cls.L])
        ax.set_xlabel('x')
        ax.set_ylabel('y')
        circles = list()
        x = list()
        y = list()
        for particle in cls.particle_list:
            x.append(particle.x)
            y.append(particle.y)
            circles.append(plt.Circle((particle.x,particle.y), radius=particle.r, color='k'))
        c = plt_circles(circles)
        ax.add_collection(c)

```

```python
N = 100
for i in range(N):
    Particles(r=0.01)

```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation

%matplotlib notebook

fig, ax = plt.subplots()

x = np.linspace(0, 10, 100)
y = np.sin(x)
line, = ax.plot(x, y)

def update_plot(frame):
    line.set_ydata(np.sin(x + frame / 10.0))
    fig.canvas.draw()

frame_count = 100
interval = 100

animation = FuncAnimation(fig, update_plot, frames=frame_count, interval=interval)

plt.show()
```
